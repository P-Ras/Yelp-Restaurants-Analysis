```python
cuisines_list = cuisines.split(',')
cuisines_list = [cuisine.strip() for cuisine in cuisines_list]

cuisines_list[0:5]
```

```python
def clean_cusine_string(cusine):
    cusine.replace(',','')
    return cusine.strip()
```

```python
def get_cusine(df_column, cuisines_list):
    cusines = [clean_cusine_string(x) for x in df_column.split(',')]
    is_cusine_found = [x in cuisines_list for x in cusines]
    if any(is_cusine_found) is True:
        index = is_cusine_found.index(True)
        return clean_cusine_string(df_column.split(',')[index])
    else:
        return 'Not Defined'
```

```python
df['Cursine'] = df.categories.apply(lambda x: get_cusine(x, cuisines_list))
print(df['Cursine'])
```
